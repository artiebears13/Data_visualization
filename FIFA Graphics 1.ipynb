```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('CompleteDataset.csv')
font_title={'fontsize':30, 'color':'black', 'family':'Times New Roman'}
axis_font={'fontsize':20, 'color':'black', 'family':'Times New Roman'}
font_title2={'fontsize':20, 'color':'black', 'family':'Times New Roman'}
```

```python
df.head()
```

# 1. Данная гистограмма покажет распределение игроков по возрасту

```python
df.Age.hist(figsize=(10,5), color='red',bins=35)
plt.title('Гистограмма возраста игроков', fontdict=font_title)
plt.xlabel('Возраст (лет)', loc='right',fontdict=axis_font)
plt.ylabel('Количество игроков, шт',fontdict=axis_font)
```

# 2. Следующий график покажет зависимость максимального, минимального и среднего рейтингов от возраста

```python
fig,ax=plt.subplots(1,figsize=(10,10))
df_mean=df.groupby(['Age'])['Overall'].mean()
df_max=df.groupby(['Age'])['Overall'].max()
df_min=df.groupby(['Age'])['Overall'].min()
plt.plot(df_mean, label='Средний рейтинг')
plt.plot(df_max,label='Максимальный рейтинг')
plt.plot(df_min,label='Минимальный рейтинг')
plt.xlabel('Возраст')
plt.ylabel('Рейтинг')
plt.legend()
plt.title('Зависимость рейтингов от возраста', fontdict=font_title)
```

# 2.1 Разобьем предыдущий график на 3 для более детального изучения

```python
fig,ax = plt.subplots(3,figsize=(10,20))
ax[0].plot(df_max)
ax[0].set_title('Распредение максимального рейтинга (Overall) по возрасту', fontdict=font_title2)

ax[1].plot(df_mean)
ax[1].set_title('Распредение среднего рейтинга (Overall) по возрасту', fontdict=font_title2)
ax[2].plot(df_min)
ax[2].set_title('Распредение минимального рейтинга (Overall) по возрасту', fontdict=font_title2)

for ax in ax.flat:
    ax.set(xlabel='возраст', ylabel='Рейтинг')
   
```
